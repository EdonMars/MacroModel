```python
predict_variable = 'XLP% n6m'

model_df = full_df.loc[:, [predict_variable, 'interest_rate', 'cpi_yoy%', 'cpi_mom%', 'gdp_qoq%', 'm2_yoy%', 'm2_mom%', 'ppi_comm_yoy%', 'ind_prod_yoy%', 'ind_prod_mom%', 'bank_prime_loan_rate', 'ppi_yoy_chem%', 'ppi_yoy_constr%', 'ppi_yoy_copper%', 'ppi_yoy_industrial%', 'ppi_yoy_metals%', 'ppi_yoy_wood%', 'ppi_yoy_proc_goods%', 'ppi_yoy_unproc_goods%', 'ppi_yoy_rubber_plastic%', 'ppi_yoy_industry%', 'truck_sales_yoy%', 'truck_sales_mom%']]
model_df.dropna(inplace=True)
X = model_df.loc[:, ['interest_rate', 'cpi_yoy%', 'cpi_mom%', 'gdp_qoq%', 'm2_yoy%', 'm2_mom%', 'ppi_comm_yoy%', 'ind_prod_yoy%', 'ind_prod_mom%', 'bank_prime_loan_rate', 'ppi_yoy_chem%', 'ppi_yoy_constr%', 'ppi_yoy_copper%', 'ppi_yoy_industrial%', 'ppi_yoy_metals%', 'ppi_yoy_wood%', 'ppi_yoy_proc_goods%', 'ppi_yoy_unproc_goods%', 'ppi_yoy_rubber_plastic%', 'ppi_yoy_industry%', 'truck_sales_yoy%', 'truck_sales_mom%']]
y = model_df[predict_variable]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

r2 = model.score(X_test, y_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print('R² score:', r2)
print('MSE:', mse)
print('MAE:', mae)
```

MORE MODELS

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso

models = [LinearRegression(), Ridge(), Lasso()]

from sklearn.model_selection import cross_val_score

for model in models:
    scores = cross_val_score(model, X_train, y_train, cv=10, scoring='r2')
    print(f"{model.__class__.__name__}: {scores.mean():.4f} ({scores.std():.4f})")
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

param_grid = {'n_neighbors': [3, 5, 7, 9, 11]}

knn = KNeighborsRegressor()
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='r2')
grid_search.fit(X_train, y_train)
best_neighbors = grid_search.best_params_['n_neighbors']

from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

model = KNeighborsRegressor(n_neighbors=best_neighbors)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("R2 score:", r2)
print("MSE:", mse)
print("MAE:", mae)
```
